```python
import pandas as pd
import numpy as np
import datetime
from datetime import date
Data=pd.read_csv('/content/drive/MyDrive/python/project python.csv')
Data
```

```python
data1=Data.groupby(['customer ID', 'Date']).agg(max=('Brand count', 'max'),total=('Brand count', 'sum')).reset_index()
data1
```

```python
data1['Percentage cutoff']= (data1['max']/data1['total'])*100
data1
```

```python
Data
```

```python
data2=Data.groupby(['customer ID', 'Date']).agg(min=('Brand count', 'min'),total=('Brand count', 'sum')).reset_index()
data2
```

```python
data2['Percentage cutoff']= (data2['min']/data2['total'])*100
data2
```

```python
Final = pd.concat([data1 , data2], axis=1, sort=False)
Final
```
